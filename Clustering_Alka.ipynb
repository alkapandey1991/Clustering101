```python
Q1 = grouped_df["TotalValue"].quantile(0.05)
Q3 = grouped_df["TotalValue"].quantile(0.95)

IQR = Q3-Q1

grouped_df = grouped_df[(grouped_df["TotalValue"]>=Q1-1.5*IQR) & (grouped_df["TotalValue"]<=Q3+1.5*IQR)]

Q1 = grouped_df["Recency"].quantile(0.05)
Q3 = grouped_df["Recency"].quantile(0.95)

IQR = Q3-Q1

grouped_df = grouped_df[(grouped_df["Recency"]>=Q1-1.5*IQR) & (grouped_df["Recency"]<=Q3+1.5*IQR)]

Q1 = grouped_df["Frequency"].quantile(0.05)
Q3 = grouped_df["Frequency"].quantile(0.95)

IQR = Q3-Q1

grouped_df = grouped_df[(grouped_df["Frequency"]>=Q1-1.5*IQR) & (grouped_df["Frequency"]<=Q3+1.5*IQR)]
```

```python
df = grouped_df[["Recency","Frequency","TotalValue"]]
df
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import sklearn
from sklearn.cluster import KMeans
import scipy
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import cut_tree
```

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
```

```python
df_scaled
```

```python
df_scaled_df = pd.DataFrame(df_scaled)
df_scaled_df
```

```python
df_scaled_df.columns = ["Recency","Frequency","Monetary"]
df_scaled_df
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(df_scaled_df)
```

```python
kmeans.labels_
```

```python
kmeans.inertia_
```

```python
#Elbow method

wss = []
n_clusters = [2,3,4,5,6,7,8,9,10]

for c in n_clusters:
    kmeans = KMeans(n_clusters= c)
    kmeans.fit(df_scaled_df)
    wss.append(kmeans.inertia_)
    
```

```python
wss
```

```python
plt.plot(wss)
```

```python
#Silhouette Score Method

n_clusters = [2,3,4,5,6,7,8,9,10]

for c in n_clusters:
    kmeans = KMeans(n_clusters= c)
    kmeans.fit(df_scaled_df)
    
    clus_label = kmeans.labels_
    
    sil = silhouette_score(df_scaled_df,clus_label)
    print("For number of clusters = {}, The Silhouette Score is: {}".format(c,sil))
    
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(df_scaled_df)
kmeans.labels_
```

```python
kmeans.inertia_
```

```python
mergings = linkage(df_scaled_df,method="single",metric = "euclidean")
```

```python
dendrogram(mergings)
plt.show()
```

```python
mergings = linkage(df_scaled_df,method="complete",metric = "euclidean")
dendrogram(mergings)
plt.show()
```

```python
hie = cut_tree(mergings,n_clusters = 3)
hie
```

```python
grouped_df["Hie_label"] = hie
grouped_df
```

```python
sns.boxplot(x="Hie_label",y="Frequency",data = grouped_df)
plt.show()
```
